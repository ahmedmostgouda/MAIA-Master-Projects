#Preprocessing the Dermo challange (Circular Windowing)

```
def create_circular_mask(h, w, center=None, radius=None):
    if center is None: # use the middle of the image
        center = [int(w/2), int(h/2)]
    if radius is None: # use the smallest distance between the center and image walls
        radius = min(center[0], center[1], w-center[0], h-center[1])
    Y, X = np.ogrid[:h, :w]
    dist_from_center = np.sqrt((X - center[0])**2 + (Y-center[1])**2)
    mask = dist_from_center <= radius
    return mask
```

##Training Set

```
from google.colab import drive
drive.mount('/content/drive')

import os
import cv2

dirs_all = os.listdir(DERMO_TRAIN_PATH)
dirs_all.sort()


for nf, folder in enumerate(dirs_all):
  print("Processing Images...")
  INTERNAL_PATH = os.path.join(DERMO_TRAIN_PATH,folder)
  all_images_path = os.listdir(INTERNAL_PATH)
  for n, file in enumerate(all_images_path):
    #print(TRAIN_PATH +"/"+folder+"/"+file) 
    img=cv2.imread(DERMO_TRAIN_PATH +"/"+folder+"/"+file)
    mask_img = create_circular_mask(img.shape[0],img.shape[1])  #Generate the circular mask
    img[~mask_img] = 0

    crop_img = img[:, 75:525]
    #plt.imshow(cv2.cvtColor(crop_img, cv2.COLOR_BGR2RGB))
    #plt.show()
    save_path = os.path.join(DERMO_TRAIN_SAVE_PATH,folder,file)
    print(save_path)
    cv2.imwrite(save_path, crop_img) 

```
